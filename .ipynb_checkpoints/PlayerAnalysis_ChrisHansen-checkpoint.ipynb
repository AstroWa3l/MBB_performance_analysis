```python
# A stat used in economics that I found from nbastuffer.com and found it interesting to interpret

def HHI_Pts(ser,Tm_Pts):
    Pts = ser.to_numpy()
    Pts = Pts[:-1]
    output = []
    for i in Pts:
        hhi_vals = ((i/Tm_Pts)**2)
        output.append(hhi_vals)
    return sum(output)
    
    
    
```

- I want to calculate the ORtg and DRtg for each individual player by applying Dean Oliver's calculations for Individual ORtg and DRtg he wrote about in his book "Basketball on Paper". I will be writing up my own short reasoning for using this statistic and how it is benefitial for determining Individual player rating (per 100 possions). For more information on the reasons for the used calculations and the nitty gritty of the algorithm can be found in these places: https://www.basketball-reference.com/about/ratings.html, https://squared2020.com/2017/10/04/deep-dive-with-python-offensive-ratings/

```python
qAST = ((Boxdata_Nototals.Min/(Tm_min/5))*(1.14*((Tm_AST - Boxdata_Nototals.Ast)/TM_FGM)))+ ((((Tm_AST/Tm_min)*Boxdata_Nototals.Min * 5 - Boxdata_Nototals.Ast) / ((TM_FGM/Tm_min)*Boxdata_Nototals.Min*5-Boxdata_Nototals.FGM))*(1-(Boxdata_Nototals.Min/(Tm_min/5))))
```

```python
FG_Part = Boxdata_Nototals.FGM * (1 - 0.5 * ((Boxdata_Nototals.Pts - Boxdata_Nototals.FTM) / (2 * Boxdata_Nototals.FGA)) * qAST)
```

```python
AST_Part = 0.5 * (((Tm_PTS - Tm_FTM) - (Boxdata_Nototals.Pts - Boxdata_Nototals.FTM)) / (2 * (Tm_FGA - Boxdata_Nototals.FGA))) * Boxdata_Nototals.Ast
```

```python
FT_Part = (1.0-(1.0-(Boxdata_Nototals.FTM / Boxdata_Nototals.FTA))**2)*0.4*Boxdata_Nototals.FTA
```

```python
Team_Scoring_Poss = TM_FGM + (1 - (1 - (Tm_FTM / Tm_FTA))**2) * Tm_FTA * 0.4
```

```python
Tm_Play = Team_Scoring_Poss/ (Boxdata.FGA + 0.4* Boxdata.FTA + Boxdata.TO)
```

```python
Opponent_TRB = float(279)
Opponent_ORB = float(72)
```

```python
Team_ORB = Tm_ORB / (Tm_ORB + (Opponent_TRB - Opponent_ORB))
```

```python
Team_ORB_Weight = ((1 - Team_ORB) * Tm_Play) / ((1 - Team_ORB) * Tm_Play + Team_ORB * (1 - Tm_Play))
```

```python
ORB_Part = Boxdata_Nototals.OREB * Team_ORB_Weight * Tm_Play
```

```python
FGxPoss = (Boxdata_Nototals.FGA - Boxdata_Nototals.FGM) * (1 - 1.07 * Team_ORB)
```

```python
FTxPoss = ((1 - (Boxdata_Nototals.FTM / Boxdata_Nototals.FTA))**2) * 0.4 * Boxdata_Nototals.FTA
```

```python
ScPoss = (FG_Part + AST_Part + FT_Part) * (1 - (Team_ORB / Team_Scoring_Poss) * Team_ORB_Weight * Tm_Play) + ORB_Part
```

```python
TotPoss = ScPoss + FGxPoss + FTxPoss + Tm_TOV
```

```python
PProd_FG_Part = 2 * (Boxdata_Nototals.FGM + 0.5 * Boxdata_Nototals.FGM3) * (1 - 0.5 * ((Boxdata_Nototals.Pts - Boxdata_Nototals.FTM) / (2 * Boxdata_Nototals.FGA)) * qAST)
```

```python
PProd_AST_Part = 2 * ((TM_FGM - Boxdata_Nototals.FGM + 0.5 * (Tm_3PM - Boxdata_Nototals.FGM3)) / (TM_FGM - Boxdata_Nototals.FGM)) * 0.5 * (((Tm_PTS - Tm_FTM) - (Boxdata_Nototals.Pts - Boxdata_Nototals.FTM)) / (2 * (Tm_FGA - Boxdata_Nototals.FGA))) * Boxdata_Nototals.Ast
```

```python
PProd_ORB_Part = Boxdata_Nototals.OREB * Team_ORB_Weight * Tm_Play * (Tm_PTS / (TM_FGM + (1 - (1 - (Tm_FTM / Tm_FTA))**2) * 0.4 * Tm_FTA))
```

```python
PProd = (PProd_FG_Part + PProd_AST_Part + Boxdata.FTM) * (1 - (Tm_ORB / Team_Scoring_Poss) * Team_ORB_Weight * Tm_Play) + PProd_ORB_Part
```

```python
ORtg = 100 * (PProd / TotPoss)
```

```python
# Add ORtg to Boxdata and need to determine what this is trying to tell me :)
Boxdata['ORtg'] =  ORtg
Boxdata
```
